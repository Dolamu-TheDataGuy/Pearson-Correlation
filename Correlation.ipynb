***Pearson correlation is a statistical method for determing the relationship between features of a dataset, the values of the pearson from -1 to 1, the correlation can either be positive or negative. When the correlation is positive, that means there is a direct relationship between the two features(i.e as one increases, the other increases also), while if it is negative, there is an inverse relationship between the two features (i.e. as one increases the other decreases).***

# IMPORTING THE DATA

***This data is a Netflix dataset that includes variety of movies and their different features like Year, Age Limit, Directors, Genre, Country e.t.c***

```python
movies = pd.read_csv('Movies_Netflix.csv')
```

```python
movies
```

```python
movies['Rotten Tomatoes'] = movies['Rotten Tomatoes'].str.replace('%', '').astype(float)
```

```python
#dropping the type column from the dataset
movies.drop('Type', inplace=True, axis = 1)
```

```python
movies
```

# OUTPING THE CORRELATION VALUES FOR EACH ATTRIBUTE

```python
correlation = movies.corr(method='pearson')
```

```python
print(correlation)
```

```python
print(correlation['Year'])
```

# Visualising the correlation using heatmap

```python
fig, ax = plt.subplots(figsize = (10, 8))
sns.heatmap(correlation, center = 0, cmap = "Blues", annot = True)
ax.set_title('Multi-Collinearity of Movies_Netflix Attributes', font = 'Century Gothic')
```
